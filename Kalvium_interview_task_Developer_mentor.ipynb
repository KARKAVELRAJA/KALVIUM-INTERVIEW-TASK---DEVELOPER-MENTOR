## **Build a image resize app. The user upload an image and selects width and height. The server resizes the image to that size, and the resized image can be viewed or downloaded by the user.**

## ***Importing Necessary Libraries***

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## ***Uploading an Image by the User***

```python
img  = mpimg.imread('/content/Karkavelraja J.jpg')
```

## ***Displaying the Original User Uploaded Image***

```python
plt.title('Original_Image')
plt.imshow(img)
plt.show()
```

## ***Function to Resizes the Original Image to the User Specified Size***

```python
def resize_fn(width, height):
  img_ = cv2.resize(img,(width,height))
  return img_
```

## ***Selecting the width and height for resizing the Image by the User***

```python
width_, height_ = input('Enter the width and height of the image that you want to resize :').split(',')
width_  = int(width_)
height_ = int(height_)
img = resize_fn(width_, height_)
```

## ***Viewing the Resized Image by the User***

```python
plt.title("Resized_Image")
plt.imshow(img)
plt.show()
```

# ***Downloading the Resized Image by the User***

```python
mpimg.imsave('resized_image.jpg',img)
```
